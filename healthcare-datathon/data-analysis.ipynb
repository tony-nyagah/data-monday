```python
import geopy
from geopy.geocoders import Nominatim
import folium

# Initialize geocoder
geolocator = Nominatim(user_agent="your_app_name")

# ... (Read your health facility data into a DataFrame)

# Geocode facility addresses
for index, row in facilities.iterrows():
    location = geolocator.geocode(
        row["Facility Name"] + ", " + row["Sub Location"] + ", Nairobi, Kenya"
    )
    if location:
        facilities.at[index, "Latitude"] = location.latitude
        facilities.at[index, "Longitude"] = location.longitude

# Create a map centered on Nairobi
map_center = [-1.2921, 36.8219]  # Latitude, Longitude of Nairobi
m = folium.Map(location=map_center, zoom_start=12)

# Add hospital markers to the map
for index, row in facilities.iterrows():
    if not pd.isna(row["Latitude"]) and not pd.isna(row["Longitude"]):
        folium.Marker(
            [row["Latitude"], row["Longitude"]], popup=row["Facility Name"]
        ).add_to(m)

# Save the map as an HTML file
m.save("nairobi_hospitals.html")
```
